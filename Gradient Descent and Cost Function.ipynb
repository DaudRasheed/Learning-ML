```python
import numpy as np
```

```python
def gradient_decent(X,Y):
    initial_m = initial_b =0
    n = len(X)
    learning_rate = 0.01
    iterations  = 1000
    for i in range(iterations):
        y_predicted = initial_m * X + initial_b
        cost = 1/n *sum((Y-y_predicted))**2
        m = -(2/n)*sum(X*(Y-y_predicted))
        b = -(2/n)*sum(Y-y_predicted)
        initial_m = initial_m - learning_rate * m
        initial_b = initial_b - learning_rate * b
        print("m = {}, b = {}, iteration  = {}, cost = {}".format(initial_m,initial_b,i, cost))
```

```python
X = np.array([1,2,3,4,5])
Y = np.array([5,7,8,9,10])
```

```python
gradient_decent(X,Y)
```
