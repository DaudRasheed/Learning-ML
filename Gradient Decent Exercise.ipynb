```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('GDexercise.csv')
```

```python
y = df.cs.values
```

```python
def gradient_decent(x,y):
    initial_m = 0
    initial_b = 0
    n = len(x)
    iteration = 100
    learning_rate = 0.001
    for i in range(iteration):
        y_predicted = initial_m * x + initial_b
        cost = (1/n) * sum((y - y_predicted))**2
        m = -(2/n) *sum(x*(y-y_predicted))
        b = -(2/n) *sum((y-y_predicted))
        initial_m = initial_m - learning_rate * m
        initial_b = initial_m - learning_rate * m
        print("m = {}, b = {}, cost = {}, iteration = {}".format(initial_m, initial_b,cost,i))
```

```python
x = df.math.values
y = df.cs.values
```

```python
gradient_decent(x,y)
```
