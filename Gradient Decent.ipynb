## Gradient Decent to find parameters

```python
def gradient_decent(x,y):
    m_curr = b_curr = 0
    iteration = 10000
    
    learning_rate = 0.005
    n = len(x)
    for i in range(iteration):
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum((y-y_predicted)**2)
        m = -(2/n) * sum(x * (y - y_predicted)) 
        b = -(2/n) * sum((y - y_predicted))
        m_curr = m_curr - learning_rate * m
        b_curr = b_curr - learning_rate * b
        print("m = {}, b = {}, cost = {}, iteration = {}".format(m_curr,b_curr,cost,i+1))
        
```

```python
x = df.Semester.values
y = df.GPA.values
```

## Values Of m and b

```python
gradient_decent(x,y)
```
